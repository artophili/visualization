```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotnine as pt
```

```python
from plotnine import ggplot, aes, geom_point

#ggplot framework
#(ggplot(data, aes(x='x_var', y='y_var')) + geom_point())
```

```python
df = pd.read_csv('C:/Users/Artophilic/Datascience Bootcamp/Visualization/iris.csv')
```

```python
df.head()
```

```python
#Data component : Loading the dataset in ggplot

pt.ggplot(df)
#This will give us a blank output as we have not specified the other two main components
```

```python
#Aesthetics : This step involves defining which variables from the dataset correspond to 
# x axis and y axis

pt.ggplot(df) + pt.aes(x='species',y='sepal_length')
```

```python
'''In the above plot, on x axis the species are plotted and on y axis sepal len is plotted 
but there is figure in the plot because we haven't declare 3rd component which is geometric objectsS'''
```

```python
pt.ggplot(df) + pt.aes(x='species',y='sepal_length') + pt.geom_col()
```

```python
#Geom helps to define various plots such as scatter plots, bar charts, or histograms etc
```

```python
#We can plot the charts using plotnine
#we just need to define the data component first then the aesthetics of
#the plot mean x and y axis. Then we have to decide which plot we need to use
#like bar, line, hist using geometric objects (geom component)
```

```python
pt.ggplot(df) + pt.aes(x='sepal_length') + pt.geom_histogram()
```

```python
pt.ggplot(df) + pt.aes(x='species',y='sepal_width') + pt.geom_point()
```

```python
pt.ggplot(df) + pt.aes(x='species',y='sepal_width') + pt.geom_boxplot()
```

```python
pt.ggplot(df) + pt.aes(x='species',y='sepal_width') + pt.geom_line()
```

```python
'''
Apart from these main 3 components, we have other optional components
Facets allow data to plot subsets of data
Statistical transformations compute the data before plotting it.
Coordinates define the position of the object in a 2D plane.
Themes define the presentation of the data such as font, color, etc.'''
```

```python
# Facets
df1 = pd.read_csv('C:/Users/Artophilic/Datascience Bootcamp/Visualization/tips.csv')
```

```python
df1.head()
```

```python
#What was the total bill according to the gender and on each day

(
    pt.ggplot(df1)
    + pt.aes(x="day", y="total_bill")
    + pt.labs(
        x="day",
        y="total_bill"
    )
    + pt.geom_col()
)
```

```python
(
    pt.ggplot(df1)
    + pt.facet_grid("sex")
    + pt.aes(x="day", y="total_bill")
    + pt.labs(
        x="day",
        y="total_bill"
    )
    + pt.geom_col()
)
```

```python
(
    pt.ggplot(df1)
    + pt.facet_grid("sex")
    + pt.aes(x="day", y="tip")
    + pt.labs(
        x="day",
        y="tip"
    )
    + pt.geom_col()
)
```

```python
#Statistical transformation

(
    pt.ggplot(df)
    + pt.aes(x='sepal_length')
    + pt.geom_histogram(bins=15)
)
```

```python
#Coordinates : It allows us to define the coordinate
(
    pt.ggplot(df)
    + pt.aes(x='sepal_length')
    + pt.geom_histogram(bins=15)
    + pt.coord_flip()
)
```

```python
#Themes

(
    pt.ggplot(df1)
    + pt.facet_grid("sex")
    + pt.aes(x="day", y="total_bill")
    + pt.labs(
        x="day",
        y="total_bill"
    )
    + pt.geom_col()
    + pt.theme_538()
)
```

```python
#Plotting multidimensional data

plot = (
    pt.ggplot(df1)
    + pt.facet_grid("sex")
    + pt.aes(x="day", y="total_bill", fill="time")
    + pt.labs(
        x="day",
        y="total_bill"
    )
    + pt.geom_col()
    + pt.theme_xkcd()
)
```

```python
plot.save("gfg plotnine tutorial.png")
```
